```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy
import scipy.special
from PIL import Image

from skimage.transform import rotate
from sklearn.neighbors import KDTree
from scipy.spatial import distance

import torch

np.seterr(all='raise')
scipy.special.seterr(all='raise')

#import cdlib
```

```python
#if centroid_list prev calculated
image_path = 'cl5-1111-small.JPG'
img = np.array(Image.open(image_path))

import pickle

centroid_list = []
with open('centroid_list.data', 'rb') as filehandle:
    centroid_list = pickle.load(filehandle)
```

```python
#if running from instances only
image_path = 'cl5-1111-small.JPG'
img = np.array(Image.open(image_path))

#load model output into instances dict: 
instances = {}
with open('instances_small.data', 'rb') as filehandle:
#with open('instances_large.data', 'rb') as filehandle:
    instances = pickle.load(filehandle)
    
'''   "pred_boxes" "pred_masks" "scores" "classes"  "image_size" '''

centroid_list = [ (((box[0]+box[2])/2).item() , ((box[1]+box[3])/2).item()) for box in instances["pred_boxes"] ]
#print(len(centroid_list))
centroid_list = np.array(centroid_list)
```

```python
len(centroid_list)
```

```python
import math

(width,height,channels) = img.shape

tree = []
tree = KDTree(centroid_list, leaf_size=10)

print(width, height)

step = int(math.sqrt(width*height)/50) #step 

query_points = np.zeros(int(width/step)*int(height/step)*2).reshape(int(width/step)*int(height/step),2)


print('step = ', step)
for i in range (0,width,step):
    #print(int(i/step)*int(width/step)+int(j/step))
    for j in range (0, height, step):
        query_points[int(i/step)*int(width/step)+int(j/step)][0] = i
        query_points[int(i/step)*int(width/step)+int(j/step)][1] = j

#wide_list = tree.query_radius(query_points, 50)
numbers_list = tree.kernel_density(query_points, h = 20, kernel = 'gaussian') #h = kernal bandwidth parameter - important**
numbers_list = np.array(numbers_list)
```

```python
numbers_list = np.transpose(numbers_list.reshape(int(width/step), int(height/step)))
```

```python
fig, ax = plt.subplots()
ax.imshow(img, cmap=plt.cm.gray)
plt.show()
plt.imshow(numbers_list)
plt.savefig('out.png', bbox_inches='tight', dpi = 300)
plt.show()
```
