```python
from sklearn import preprocessing

combinedarr = combined.values
min_max_scaler = preprocessing.MinMaxScaler()
combinedarr = min_max_scaler.fit_transform(combinedarr)
pd.DataFrame(combinedarr)
```

```python
agg = [sum(i[29:35])/6 for i in combinedarr]
agg = min_max_scaler.fit_transform(np.array(agg).reshape(-1,1))
agg
```

```python
plt.plot(sorted(agg))
```

```python
newarr = [i[0:23] for i in combinedarr]

len(newarr[0])
```

```python
for i,ar in enumerate(newarr):
    newarr[i] = np.append(ar,agg[i])
```

```python
np.array(newarr).corr()
```

```python
cm = pd.DataFrame(newarr).corr()
```
